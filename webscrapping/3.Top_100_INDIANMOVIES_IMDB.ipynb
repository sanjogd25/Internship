```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
page
```

```python
## display the page source code
```

```python
soup = BeautifulSoup(page.content, "html.parser")
soup
```

```python
scraped_movies = soup.find_all('td',class_="titleColumn")
scraped_movies
```

```python
movies = []
for movie in scraped_movies:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies
```

```python
scraped_ratings = soup.find_all('td',class_="ratingColumn imdbRating")
scraped_ratings
```

```python
ratings = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', "")
    ratings.append(rating)
ratings
```

```python
scraped_years = soup.find_all('span',class_="secondaryInfo")
scraped_years
```

```python
years = []
for year in scraped_years:
    year = year.get_text().replace('/span>,', "")
    years.append(year)
years
```

```python
import pandas as pd

data = pd.DataFrame()
data['Movie names'] = movies
data['Ratings'] = ratings
data['Year of release'] = years
data.head()
```
